```python
grupoestavel = np.random.normal(loc=0,scale=5, size=100)
grupovolatil = np.random.normal(loc=0, scale=30, size=100)
```

```python
plt.figure(figsize=(12,6))

sns.histplot(grupoestavel, kde=True, color="blue", label="Grupo Estável", alpha=0.5, bins=50)
sns.histplot(grupovolatil, kde=True, color="red", label="Grupo Volátil", alpha=0.5, bins=50)
plt.title("Comparação de Dispersão: Mesma Média, Riscos Diferentes")
plt.legend()
plt.show()
```

```python
print("volatil " ,  grupovolatil.std()/grupovolatil.mean())
print("estavel ", grupoestavel.std()/grupoestavel.mean())
```

```python
premios = np.array([0,1,1,2,20,20,20,20,30,60])
custo = 20

saldo_inicial=200
premios.mean()

dinheiro=saldo_inicial
rodadas=0
historico=[dinheiro]
while dinheiro >=custo :
  rodadas+=1
  ganho = np.random.choice(premios)
  dinheiro+= ganho -custo
  historico.append(dinheiro)
  #if rodadas <=10:
    #print(f"Rodada {rodadas}: Ganhou R$ {ganho}. Saldo atual: R$ {dinheiro}")
plt.plot(historico, marker='o', linestyle='-')
plt.xlabel('Rodada')
plt.ylabel('Saldo')
plt.title('Evolução do Saldo ao Longo das Rodadas')
```
