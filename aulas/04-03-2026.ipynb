```python
class Cliente:
    def __init__(self, nome: str, telefone: str):
        self.nome = nome
        self.telefone = telefone

    def __repr__(self) -> str:
        return f"""
        === Dados do Cliente ===
        Nome: {self.nome}
        Telefone: {self.telefone}
                """ 

ana = Cliente('Ana', '97797-2067')
ana
```

```python
class Dispositivo:
    def __init__(self, nome: str, marca: str, sintoma: str):
        self.nome = nome
        self.marca = marca
        self.sintoma = sintoma

    def testar(self) -> str:
        return f"""
        === Identificando características do Dispositivo ===
        O {self.nome} da marca {self.marca} {self.sintoma}
                """ 

    def __repr__(self) -> str:
        return f"""
        === Dados ===
        Nome: {self.nome}
        Telefone: {self.marca}
        Sintoma: {self.sintoma}
                """ 

d = Dispositivo('Carregador do Além', 'Genérica', 'só carrega em 37ºC')
print(d)
print(d.testar())
```

```python
class OrdemDeServico:
    def __init__(self, cliente: Cliente, dispositivo: str):
        self.cliente = cliente
        self.dispositivo = dispositivo
        self._status = "ABERTA"
        self.diagnostico = None

    @property # isso faz o código achar que um método é um atributo 
    def status(self) -> str:
        return self._status

    def registrar_diagnostico(self, texto: str) -> None:
        self.diagnostico = texto
    
    def concluir(self) -> None:
        if self.diagnostico is None:
            raise ValueError('Não tem diagnóstico')
        self._status = "CONCLUIDA" 

    def resumo(self) -> str:
        return (
            f"OS | {self._status} | Cliente: {self.cliente} |"
            f"Dispositivo {self.dispositivo.nome}"
        )       
```

```python
ana = Cliente('Ana', '11 91919-1919')
carregar = Dispositivo('Cabo sem fio', 'Genérico', 'Não passa cabo nem fio')
os1 = OrdemDeServico(ana, carregar)
os1.resumo()
```

```python
try:
    os1.concluir()
except ValueError as e:
    print(e)
```

```python
os1.registrar_diagnostico('Bluetooth')
os1.concluir()
os1.resumo()
```
