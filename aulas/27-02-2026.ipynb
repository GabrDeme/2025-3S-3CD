```python
class Cliente:
    def __init__(self, nome: str, telefone: str):
        self.nome = nome
        self.telefone = telefone

    def __repr__(self) -> str:
        return f"""
        === Dados do Cliente ===
        Nome: {self.nome}
        Telefone: {self.telefone}
                """ 

ana = Cliente('Ana', '97797-2067')
ana
```

```python
class Dispositivo:
    def __init__(self, nome: str, marca: str, sintoma: str):
        self.nome = nome
        self.marca = marca
        self.sintoma = sintoma

    def testar(self) -> str:
        return f"""
        === Identificando características do Dispositivo ===
        O {self.nome} da marca {self.marca} {self.sintoma}
                """ 

    def __repr__(self) -> str:
        return f"""
        === Dados ===
        Nome: {self.nome}
        Telefone: {self.marca}
        Sintoma: {self.sintoma}
                """ 

d = Dispositivo('Carregador do Além', 'Genérica', 'só carrega em 37ºC')
print(d)
print(d.testar())
```
