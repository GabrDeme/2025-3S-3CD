```python
arquivos_mb = []
arquivos_gb = []
arquivos_tb = []

arquivos_recebidos = []
```

```python
def organizar(lista):
  for classificacao in lista:
    if classificacao < 1024:
      arquivos_mb.append(classificacao) # não ta adicionando nada porque o append está vazio
    elif classificacao < 1048576:
      arquivos_gb.append(classificacao) # não ta adicionando nada porque o append está vazio
    else:
      arquivos_tb.append(classificacao) # não ta adicionando nada porque o append está vazio
```

```python
print("Bem-vindo ao organiza pra você! Digite os tamanhos dos arquivos (em mb, gb etc). Digite '0' para finalizar")

while True:
  arquivo = input("Quantos mb há em seus arquivos? ")
  arquivo = int(arquivo)
  if arquivo == 0:
    break
  else: # precisei colocar um else
    arquivos_recebidos.append(arquivo)

organizar(arquivos_recebidos)

# coloquei um print para mostrar o que está sendo feito
print(f"""

MB: {arquivos_mb}
GB: {arquivos_gb}
TB: {arquivos_tb}

""")
```
