```python
from collections import Counter
```

```python
def calcmedia(X):
    somatorio=0

    for x in X:
        somatorio=somatorio+x

    n=len(X)

    media = somatorio/n

    return media
```

```python
Y=[1, 2, 3, 2, 6, 53, 5,3, 553, 43,7]
print( calcmedia(Y) )
```

```python
def calcmediana(X):
  X = np.sort(X)

  n= X.size

# X[0] x[1] x[2] x[3] x[4] -> n=5
# X[0] x[1] x[2] x[3]  -> n=4

  if n%2==1:
    return X[ n//2 ]

  return (X[n//2]+X[n//2 - 1])/2

  # n*n   ex bubblesort
  # n*log(n)   ex quicksort

  # 10*10 =100
  # 10*log2(10)  10*3.5 = 35

  # 10^6 * 10^6 = 10^12
  # 10^6 * log2(10^6) = 2*10^7
```

```python
Y = [4, 5, 7,  1, 9, 10]

print(calcmediana(Y))
```

```python
def calcmoda(X):
  contagem = Counter(X)
  return contagem.most_common(1)[0][0]

```

```python
Y = ["4", "5", 7,  1, 3.0, int(3),int(3),int(3),int(3),int(3),int(3),int(3),int(3),int(3),int(3),int(3), 9, 10, "4", "5","4", "3", "3", "3", "3", 7, 2, 2, 5,3, 3, 3, 3, 3, 3, 3]

calcmoda(Y)
```
