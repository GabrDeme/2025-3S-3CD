```python
pop = [20, 30, 40]
pop2 = [1, 5, 7, 3, 4]

combinacoes = []
for x in pop:
    for y in pop:
        combinacoes.append((x, y))

combinacoes

comb = np.array(combinacoes)
comb

print(comb)
print("===")
print(comb.var(axis=1, ddof=1).mean())
print("===")
print(comb.var(axis=1, keepdims=True))
print("===")
print(comb.var(axis=0))
print("===")
print(comb.std(axis=1, keepdims=True, ddof=1).mean())
```

```python
grupoestavel = np.random.normal(loc=100, scale=5, size=1000)
grupovolatil = np.random.normal(loc=100, scale=30, size=100)
```

```python
plt.figure(figsize=(12, 6))

sns.histplot(grupoestavel, kde=True, color="blue", label="Grupo Estável", alpha=0.5) 
sns.histplot(grupovolatil, kde=True, color="red", label="Grupo Volátil", alpha=0.5)

plt.legend()
```

```python
print(grupoestavel.std()/grupoestavel.mean())
print(grupovolatil.std()/grupovolatil.mean())
```

```python
premios = [0,1,1,2,20,20,20,20,30,60]
custo = 20

saldo_inicial=200


dinheiro=saldo_inicial
rodadas=0
historico=[dinheiro]
while dinheiro >=custo:
  rodadas+=1
  ganho = np.random.choice(premios)
  dinheiro+= ganho-custo
  historico.append(dinheiro)
  # if rodadas <=10:
   # print(f"Rodada {rodadas}: Ganhou R$ {ganho}. Saldo atual: R$ {dinheiro}")

plt.plot(historico, marker='o', linestyle='-')
plt.xlabel("Rodada")
plt.ylabel("Saldo")
plt.title('Evolucao do Saldo ao Longo das Rodadas')
```
