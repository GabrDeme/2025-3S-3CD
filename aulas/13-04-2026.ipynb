```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import t as t_dist, norm
```

```python
x=np.linspace(-5,5,500)
normal_pdf = norm.pdf(x)

fig, axes = plt.subplots(1,2, figsize=(8, 4))
ax = axes[0]
configs = [
            (1, '#f07070', 'df=1'),
            (3, '#fa8040', 'df=3'),
            (6, '#50bf40', 'df=6'),
            (18, '#60bf40', 'df=18'),
            (100, '#000', 'df=100'),
           ]

ax.plot(x,normal_pdf)
for df,cor,label in configs:
  ax.plot(x, t_dist.pdf(x, df), cor, label=label, alpha=0.7)
ax.legend()

ax2 = axes[1]
x_zoom = np.linspace(1.5, 5, 300)
for df,cor,label in configs:
  ax2.plot(x_zoom, t_dist.pdf(x_zoom, df), cor, label=label, alpha=0.7)
```

```python
popmi = 0
popsigma = 1
n_sim = 20000
tamanhos = [3, 5, 10, 30]

cores = ['#f07070', '#fa8040', '#50bf40','#60bf40', '#000']

fig.axes = plt.subplots(2, 2, figsize=(12, 12))

x_range = np.linspace(-6, 6, 600)

for idx, (n_amostras, cor) in enumerate(zip(tamanhos, cores)):
    df = n_amostras - 1
    t_scores = []
    for _ in range(n_sim):
        amostra = np.random.normal(loc = popmi, scale = popsigma, n_amostras)
        x_bar = amostra.mean()
        s = amostra.std(ddof = 1)
        t_scores.append((x_bar - popmi) / (s/np.sqrt(n_amostras)))
    t_scores = np.array(t_scores)
    ax = axes[idx]
    ax.hist(t_scores, bins=120, density = True)
```
