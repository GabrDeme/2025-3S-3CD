```python
import nbformat
from nbformat.v4 import new_notebook, new_code_cell

def remove_comments(code: str) -> str:
    lines = []
    for line in code.split("\n"):
        stripped = line.strip()
        if stripped.startswith("#"):
            continue
        lines.append(line)
    return "\n".join(lines)

def split_into_cells(code: str):
    blocks = []
    current = []

    for line in code.split("\n"):
        if line.strip() == "" and current:
            blocks.append("\n".join(current))
            current = []
        else:
            current.append(line)

    if current:
        blocks.append("\n".join(current))

    return blocks

def create_notebook(py_file, output_file):
    with open(py_file, "r", encoding="utf-8") as f:
        code = f.read()

    code = remove_comments(code)
    blocks = split_into_cells(code)

    nb = new_notebook()

    for block in blocks:
        if block.strip():
            nb.cells.append(new_code_cell(block))

    with open(output_file, "w", encoding="utf-8") as f:
        nbformat.write(nb, f)

    print(f"Notebook criado: {output_file}")


if __name__ == "__main__":
    create_notebook("05-05-2026.py", "05-05-2026.ipynb")
```
