```python
def h(x, ordem=2):
    ret=x
    for i in range(2, ordem+1):
        ret += x**i/i*(-1)**(i+1)
    return ret
```

```python
print(np.log(1+delta))
h(delta, ordem=4)
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import scipy.stats as stats
```

```python
np.random.seed(42)
pesos_cafe = np.random.normal(loc=500, scale=4, size=1000)

df_cafe = pd.DataFrame({'Peso_g': pesos_cafe})
```

```python
df_cafe
```

```python
df_cafe.mean()
```

```python
df_cafe.std()
```

```python
fig = px.histogram(df_cafe, x='Peso_g', nbins=100)
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x=df_cafe['Peso_g']))

x_teorico = np.linspace(df_cafe['Peso_g'].min(), df_cafe['Peso_g'].max(),100)
y_teorico = stats.norm.pdf(x_teorico, loc=500, scale=3.9)

fig.add_trace(go.Scatter(x=x_teorico, y=y_teorico, mode='lines'))
fig.show()
```
