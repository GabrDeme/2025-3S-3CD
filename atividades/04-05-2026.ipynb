```python
import importlib.util

# ============================================================
# 1. CARREGAR O ARQUIVO COM NOME NÃO PADRÃO
# ============================================================
file_path = "/../28-04-2026.py"

spec = importlib.util.spec_from_file_location("engine_module", file_path)
engine_module = importlib.util.module_from_spec(spec)
spec.loader.exec_module(engine_module)

# ============================================================
# 2. ACESSAR CLASSE E FUNÇÕES DO ARQUIVO
# ============================================================
InsightCalculadoEngine = engine_module.InsightCalculadoEngine
load_json = engine_module.load_json
save_json = engine_module.save_json

# ============================================================
# 3. CARREGAR JSON DE ENTRADA
# ============================================================
entrada = load_json("04-05-2026.json")

# ============================================================
# 4. EXECUTAR ENGINE
# ============================================================
engine = InsightCalculadoEngine(entrada)

resultado_aula1 = engine.aula_1()
resultado_aula2 = engine.aula_2()
resultado_aula3 = engine.aula_3()
resultado_aula4 = engine.aula_4()

# ============================================================
# 5. SALVAR RESULTADOS EM JSON
# ============================================================
save_json("saida_aula1.json", resultado_aula1)
save_json("saida_aula2.json", resultado_aula2)
save_json("saida_aula3.json", resultado_aula3)
save_json("saida_aula4.json", resultado_aula4)

# ============================================================
# 6. GERAR RELATÓRIO .DOC (arquivo texto compatível com Word)
# ============================================================
def gerar_doc():
    with open("relatorio_final.doc", "w", encoding="utf-8") as f:
        f.write("RELATÓRIO INSIGHT CALCULADO\n\n")

        f.write("=== AULA 1 ===\n")
        f.write(str(resultado_aula1))
        f.write("\n\n")

        f.write("=== AULA 2 ===\n")
        f.write(str(resultado_aula2))
        f.write("\n\n")

        f.write("=== AULA 3 ===\n")
        f.write(str(resultado_aula3))
        f.write("\n\n")

        f.write("=== AULA 4 ===\n")
        f.write(str(resultado_aula4))
        f.write("\n\n")

        f.write("\n=== CÓDIGO UTILIZADO ===\n")
        f.write("Arquivo carregado dinamicamente com importlib\n")

gerar_doc()

print("Relatório gerado com sucesso: relatorio_final.doc")
```
