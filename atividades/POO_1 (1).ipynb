```python
class Pessoa:
    def __init__(self, nome, idade, profissao):
        self.nome = nome
        self.idade = idade
        self.profissao = profissao

    def apresentar(self):
        return f"Oi, eu sou {self.nome}, tenho {self.idade} anos e sobrevivo sendo {self.profissao}."
```

```python
class Cachorro:
    def __init__(self, nome, raca, nivel_fofura):
        self.nome = nome
        self.raca = raca
        self.nivel_fofura = nivel_fofura

    def apresentar(self):
        return f"Oi, este é o meu cachorro {self.nome}, ele é da raça {self.raca} e ele é muito fofo, mais ou menos uns {self.nivel_fofura}/10)"
```

```python
pessoa1 = Pessoa("Carlos", 28, "Programador")
cachorro1 = Cachorro("Thor", "Vira-lata Premium", 11)

print(pessoa1.apresentar())
print(cachorro1.apresentar())
```
