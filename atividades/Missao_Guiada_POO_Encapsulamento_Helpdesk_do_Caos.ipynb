
## вң… MissГЈo 3 вҖ” Classe `Chamado` (a OS do caos)

Aqui vocГӘ junta tudo: `Cliente` + `Item` + as regras do sistema.

### Regras de negГіcio
- Chamado nasce com status **ABERTO**
- DiagnГіstico comeГ§a como `None`
- NГЈo pode **concluir** sem **diagnГіstico**
- Se tentar concluir sem diagnГіstico вҶ’ `raise ValueError("...")`

### Atributos mГӯnimos
- `cliente: Cliente`
- `item: Item`
- `problema: str`
- `diagnostico: Optional[str]` (comeГ§a `None`)
- `status` (por enquanto pode ser `self.status = "ABERTO"` вҖ” depois vamos encapsular)

### MГ©todos
- `diagnosticar(texto)`
- `concluir()`
- `resumo()` вҶ’ retorna uma string bonita


```python
class Chamado:
    def __init__(self, cliente: Cliente, item: Item, problema: str):
        # TODO: salvar cliente, item e problema
        # TODO: diagnostico comeГ§a None
        # TODO: status comeГ§a "ABERTO"
        self.cliente = cliente
        self.item = item
        self.problema = problema
        self.status = "ABERTA"
        self.diagnostico = None

    def diagnosticar(self, texto: str) -> None:
        # TODO: validar texto (nГЈo pode vazio/sГі espaГ§os)
        # TODO: salvar em self.diagnostico
        if not texto:
            raise ValueError("Coloca os dados, precisamos de dados")
        self.diagnostico = texto

    def concluir(self) -> None:
        # Regra: nГЈo conclui sem diagnГіstico
        # TODO: se self.diagnostico for None -> raise ValueError("...")
        # TODO: senГЈo -> mudar status para "CONCLUIDO"
        if self.diagnostico is None:
            raise ValueError('NГЈo tem diagnГіstico')
        self.status = "CONCLUIDA" 

    def resumo(self) -> str:
        # TODO: retornar uma string com status, cliente e item
        # SugestГЈo:
        # "OS | <STATUS> | Cliente: <nome> | Item: <nome> | DiagnГіstico: <...>"
        return (
            f"OS | {self.status} | Cliente: {self.cliente.nome} | Item: {self.item.nome} | "
            f"DiagnГіstico: {self.diagnostico}"
        ) 
```


## вң… MissГЈo 6 вҖ” Encapsulamento (Parte 7)

Agora vem a regra: **ninguГ©m muda status na mГЈo**.

VocГӘ deve:
- trocar `status` por um atributo interno: `self._status`
- criar `@property def status(self): return self._status`
- **nГЈo criar setter**
- status sГі muda via mГ©todos (`concluir`, etc.)

### Tarefa
1) Atualize a classe `Chamado`:
- onde era `self.status`, virar `self._status`
- crie a property `status` (somente leitura)
- ajuste `resumo()` para usar `self.status`

2) Teste obrigatГіrio:
- tente fazer `ch1.status = "HACKEADO"`
- deve dar `AttributeError` e vocГӘ captura com `try/except`


```python
# вң… RefaГ§a a classe Chamado AQUI (substituindo a anterior)
# Dica: vocГӘ pode copiar/colar e ajustar.

class Chamado:
    def __init__(self, cliente: Cliente, item: Item, problema: str):
        # TODO: salvar cliente, item, problema
        # TODO: diagnostico comeГ§a None
        # TODO: status encapsulado comeГ§a "ABERTO" em self._status
        self.cliente = cliente
        self.item = item
        self.problema = problema
        self._status = "ABERTA"
        self.diagnostico = None

    def resumo(self) -> str:
        # TODO: retornar uma string com status, cliente e item
        # SugestГЈo:
        # "OS | <STATUS> | Cliente: <nome> | Item: <nome> | DiagnГіstico: <...>"
        return (
            f"OS | {self._status} | Cliente: {self.cliente.nome} | Item: {self.item.nome} | "
            f"DiagnГіstico: {self.diagnostico}"
        ) 


    @property
    def status(self) -> str:
        # TODO: retornar self._status
        return self._status

    def diagnosticar(self, texto: str) -> None:
        # TODO: validar e salvar diagnГіstico
        if not texto:
            raise ValueError("Coloca os dados, precisamos de dados")
        self.diagnostico = texto

    def concluir(self) -> None:
        # TODO: raise se nГЈo tiver diagnГіstico
        # TODO: mudar self._status para "CONCLUIDO"
        if self.diagnostico is None:
            raise ValueError('NГЈo tem diagnГіstico')
        self._status = "CONCLUIDA" 

    def resumo(self) -> str:
        # TODO: retornar string com self.status, cliente e item
        return (
            f"OS | {self._status} | Cliente: {self.cliente.nome} | Item: {self.item.nome} | "
            f"DiagnГіstico: {self.diagnostico}"
        )
```

### рҹ§Ә Teste final (MissГЈo 6)
Se passar aqui, vocГӘ concluiu atГ© Encapsulamento вң…

```python
# Recriar objetos (porque redefinimos a classe Chamado)
c1 = Cliente("Dona CГЎssia", "11 99999-0000")
i1 = Item("Geladeira que chora", "EletrodomГ©stico")
ch1 = Chamado(c1, i1, "Toda vez que abre a porta ela faz drama.")

print(ch1.resumo())

# 1) tentar concluir sem diagnГіstico
try:
    ch1.concluir()
except ValueError as e:
    print("Erro esperado вң…:", e)

# 2) diagnosticar e concluir
ch1.diagnosticar("PossГӯvel excesso de sentimentos no compressor.")
ch1.concluir()
print(ch1.resumo())

# 3) testar encapsulamento
try:
    ch1.status = "HACKEADO"
except AttributeError as e:
    print("Encapsulado вң…:", e)
```
