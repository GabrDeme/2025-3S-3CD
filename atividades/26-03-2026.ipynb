```python
from datetime import datetime
```

# Classes

## Classes Base

```python
class Funcionario:
    def __init__(self, id, nome, cargo):
        self.id = id
        self.nome = nome
        self.cargo = cargo

    def realizar_servico(self, agendamento):
        print(f"{self.nome} está realizando o serviço: {agendamento.servico.nome}")

    def vender_produto(self, cliente, compra):
        print(f"{self.nome} vendeu produtos para {cliente.nome}")

    def comprar_estoque(self, estoque, produto, quantidade):
        estoque.adicionar_produto(produto, quantidade)
        print(f"{self.nome} comprou {quantidade} unidades de {produto.nome} para o estoque")
```

```python
class Cliente:
    def __init__(self, id, nome, telefone):
        self.id = id
        self.nome = nome
        self.telefone = telefone

    def agendar_servico(self, servico, data):
        agendamento = Agendamento(1, data, "Pendente", self, servico)
        print(f"{self.nome} agendou o serviço: {servico.nome}")
        return agendamento

    def comprar_produto(self, produtos):
        compra = Compra(1, datetime.now())
        for produto in produtos:
            compra.adicionar_produto(produto)
        print(f"{self.nome} realizou uma compra")
        return compra
```

## Entidades Principais

```python
class Servico:
    def __init__(self, id, nome, preco, duracao):
        self.id = id
        self.nome = nome
        self.preco = preco
        self.duracao = duracao
```

```python
class Produto:
    def __init__(self, id, nome, preco, quantidade):
        self.id = id
        self.nome = nome
        self.preco = preco
        self.quantidade = quantidade
```

```python
class Agendamento:
    def __init__(self, id, data, status, cliente, servico):
        self.id = id
        self.data = data
        self.status = status
        self.cliente = cliente
        self.servico = servico

    def confirmar(self):
        self.status = "Confirmado"
        print("Agendamento confirmado!")

    def cancelar(self):
        self.status = "Cancelado"
        print("Agendamento cancelado!")
```

```python
class Pagamento:
    def __init__(self, id, valor, metodo):
        self.id = id
        self.valor = valor
        self.metodo = metodo
        self.status = "Pendente"

    def processar_pagamento(self):
        self.status = "Pago"
        print(f"Pagamento de R${self.valor} realizado via {self.metodo}")
```

```python
class Estoque:
    def __init__(self, id):
        self.id = id
        self.produtos = {}

    def adicionar_produto(self, produto, quantidade):
        if produto.nome in self.produtos:
            self.produtos[produto.nome].quantidade += quantidade
        else:
            produto.quantidade = quantidade
            self.produtos[produto.nome] = produto

    def remover_produto(self, produto_nome, quantidade):
        if produto_nome in self.produtos:
            self.produtos[produto_nome].quantidade -= quantidade
            print(f"{quantidade} unidades removidas de {produto_nome}")
```

```python
class Compra:
    def __init__(self, id, data):
        self.id = id
        self.data = data
        self.produtos = []
        self.total = 0

    def adicionar_produto(self, produto):
        self.produtos.append(produto)
        self.total += produto.preco

    def calcular_total(self):
        return self.total
```

# Rodando o código

```python
# Criando objetos
cliente = Cliente(1, "João", "99999-9999")
funcionario = Funcionario(1, "Maria", "Atendente")

servico = Servico(1, "Corte de Cabelo", 50.0, 30)
produto = Produto(1, "Shampoo", 25.0, 10)

estoque = Estoque(1)

# Fluxo de serviço
agendamento = cliente.agendar_servico(servico, datetime.now())
agendamento.confirmar()
funcionario.realizar_servico(agendamento)

pagamento_servico = Pagamento(1, servico.preco, "Pix")
pagamento_servico.processar_pagamento()

# Fluxo de compra
compra = cliente.comprar_produto([produto])
funcionario.vender_produto(cliente, compra)

pagamento_compra = Pagamento(2, compra.calcular_total(), "Cartão")
pagamento_compra.processar_pagamento()
```
