## Associação



```python
class Disciplina:
    def __init__(self, nome):
        self.nome = nome

class Professor:
    def __init__(self, nome, disciplina):
        self.nome = nome
        self.disciplina = disciplina

    def apresentar(self):
        print(
            f"""
        Professor: {self.nome}
        Disciplina dele: {self.disciplina.nome}
            """)
        
disciplina_python = Disciplina("Anacondas para Python")
professor_sucuri = Professor("Sucuri", disciplina_python)

professor_sucuri.apresentar()
```

## Herança



```python
class Funcionario:
    def __init__(self, nome, salario):
        self.nome = nome
        self.salario = salario
    def exibir_dadaos(self):
        print(
            f"""
        Funcionárior: {self.nome}
        Salário: R$ {self.salario}
            """)
        
class Professor(Funcionario): 
    def __init__(self, nome, salario, area ):
        super().__init__(nome, salario)
        self.area = area
    def exibir_dados(self):
        super().exibir_dados()
        print(f"Área: {self.area}")

prof = Professor("Mariana", 4500, "Tecnologia")
prof.exibir_dadaos()
```

```python
from abc import ABC, abstractclassmethod
```

```python
class FormaPagamento(ABC):
    @abstractclassmethod
    def pagar(self, valor):
        pass

class Pix(FormaPagamento):
    def pagar(self, valor):
        print(f"Pagamento de R$ {valor:.2f} realizado via Pix")

class Cartao(FormaPagamento):
    def pagar(self, valor):
        print(f"Pagamento de R$ {valor:.2f} realizado via Cartão")

pagamento1 = Pix()
pagamento1.pagar(10)

pagamento2 = Cartao()
pagamento2.pagar(50)
```
