### Código Gerador de Dados

```python
import numpy as np
import pandas as pd

np.random.seed(42) # Teste seeds diferentes também
transacoes_normais = np.random.normal( loc=150 , scale=30 , size=4995)
fraudes = np . array ([850 , 1200 , 950 , 800 , 750]) # Valores anomalos
todas_transacoes = np.concatenate([ transacoes_normais , fraudes ])
np.random.shuffle( todas_transacoes )
df_pix = pd.DataFrame({ 'Valor_Transacao': todas_transacoes })
print (" Dados carregados no dataframe 'df_pix'.")
```
